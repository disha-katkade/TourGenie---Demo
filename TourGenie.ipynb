# Simple Travel Recommendation

```python

import pandas as pd

# Sample dataset
data = {
    'Place': ['Heritage Fort', 'Mountain View', 'Beach Cafe', 'Cultural Street', 'Nature Park'],
    'Category': ['History', 'Adventure', 'Food', 'Culture', 'Nature']
}

df = pd.DataFrame(data)

# Function to recommend places
def recommend_place(preference):
    print(f"\nRecommended places for {preference} lovers:\n")
    results = df[df['Category'].str.contains(preference, case=False)]
    if results.empty:
        print("No direct match found! Try another preference.")
    else:
        for place in results['Place']:
            print("•", place)


recommend_place('Nature')
```

```python
recommend_place('Culture')
```

```python
recommend_place('Food')
```

# Simple Sentiment Analysis

```python

from textblob import TextBlob

# Sample user reviews
reviews = [
    "The mountain trail was amazing and peaceful!",
    "Food was terrible and overpriced.",
    "Loved the heritage fort, it was a great experience.",
    "The beach was too crowded and dirty."
]

print("\nUser Review Sentiment Analysis:\n")

for r in reviews:
    polarity = TextBlob(r).sentiment.polarity
    sentiment = "Positive" if polarity > 0 else "Negative" if polarity < 0 else "Neutral"
    print(f"Review: {r}\nSentiment: {sentiment}\n")
```
